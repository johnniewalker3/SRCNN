```python
def modcrop(image, scale):
    h = image.shape[0]
    w = image.shape[1]
    if len(image.shape) == 3:
        h = h - h % scale
        w = w - w % scale
        image = image[0:h, 0:w, :]
    else:
        h = h - h % scale
        w = w - w % scale
        image = image[0:h, 0:w]

    return image
        
```

```python
def preprocess(image, scale):
    img = modcrop(image, scale)
    #img = image[:, :]
    img = img.astype("float32")
    label = img / 255.
    sub_image = cv2.GaussianBlur(label, (3, 3), sigmaX=1, sigmaY=1).copy()
    input1 = cv2.resize(sub_image, (label.shape[1]//scale, label.shape[0]//scale), interpolation=cv2.INTER_CUBIC)
    input1 = cv2.resize(input1, (label.shape[1], label.shape[0]), interpolation=cv2.INTER_CUBIC)

    return label, input1
```

```python
def setup(path, scale, stride, img_size, label_size):
    X_train = []
    Y_train = []
    nx = 0
    ny = 0
    padding = int(abs(img_size - label_size) / 2)
    for image in os.listdir(path):
        image = cv2.imread(os.path.join(path, image), cv2.IMREAD_COLOR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
        h = image.shape[0]
        w = image.shape[1]
        h = h - h % scale
        w = w - w % scale
        image = image[0:h, 0:w, :]
        for x in range(0, h - img_size + 1, stride):
           for y in range(0, w - img_size + 1, stride):
              sub_label, sub_input = preprocess(image[x:x+img_size, y:y+img_size, 0], scale)
              sub_input = np.reshape(sub_input, (img_size, img_size, 1))
              #sub_label = sub_label[padding:-padding, padding:-padding, :]
              sub_label = np.reshape(sub_label, (label_size, label_size, 1))

              X_train.append(sub_input)
              Y_train.append(sub_label)

    X_train = np.array(X_train)
    Y_train = np.array(Y_train)
    return X_train, Y_train, [nx, ny]
```

```python
def psnr_2(target, ref):
    target = target.astype("float32")
    ref = ref.astype("float32")
    diff = target - ref
    rmse = sqrt(np.mean(diff ** 2))
    psnr = 20 * log10(255.0 / rmse)
    return psnr
```

```python
def psnr(target, ref):
    target = target.astype("float32")
    ref = ref.astype("float32")
    diff = target - ref
    rmse = sqrt(np.mean(diff ** 2))
    psnr = 20 * log10(1.0 / rmse)
    return psnr
```

```python
def shave(image, padding):
    if len(image.shape) == 3:
        image = image[padding:-padding, padding:-padding, :]
    else:
        image = image[padding:-padding, padding:-padding]

    return image
```

```python
def evaluate_in_ycrcb(obj):
    psnr_srcnn = 0
    psnr_bicubic = 0
    for file in os.listdir(set5_path):
          image = cv2.imread(os.path.join(set5_path, file))
          image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
          label, bicubic = preprocess(image[:, :, 0], obj.scale)
          bicubic = np.reshape(bicubic, (1, bicubic.shape[0], bicubic.shape[1], 1))
          out = (obj.model(bicubic)).numpy()
          out = out[0, :, :, 0]
          out = out[:, :] * 255
          out[out[:, :]>255] = 255
          out[out[:, :] < 0] = 0
          out = out.astype(np.uint8)
          out = out[obj.scale:-obj.scale, obj.scale:-obj.scale]
          label = label[:, :]*255
          label[label[:, :]>255] = 255
          label[label[:, :]<0] = 0
          label = label.astype(np.uint8)
          label = label[obj.scale:-obj.scale, obj.scale:-obj.scale]
          bicubic = np.reshape(bicubic, (bicubic.shape[1], bicubic.shape[2]))
          bicubic = bicubic[:, :] * 255
          bicubic[bicubic[:, :]>255] = 255
          bicubic[bicubic[:, :]<0] = 0
          bicubic = bicubic.astype(np.uint8)
          bicubic = bicubic[obj.scale:-obj.scale, obj.scale:-obj.scale]
          psnr_srcnn += psnr_2(label, out)
          psnr_bicubic += psnr_2(label, bicubic)
    print("Mean PSNR for epoch: %.3f" % (psnr_srcnn/len(os.listdir(set5_path))))
    return psnr_srcnn/len(os.listdir(set5_path)), psnr_bicubic/len(os.listdir(set5_path))
```

```python
def evaluate_in_rgb(srcnn, path):
    psnr_srcnn = 0
    srcnn_ssim = []
    bicubic_ssim = []
    psnr_bicubic = 0
    list1 = []
    list2 = []
    for image in os.listdir(path):
        image = cv2.imread(os.path.join(path, image), cv2.IMREAD_COLOR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

        color = image[:, :, 1:3]
        color, bad_color = preprocess(color, srcnn.scale)
        color = color[:, :, :] * 255
        color[color[:, :, :]>255] = 255
        color[color[:, :, :]<0] = 0
        color = color.astype(np.uint8)
        color = shave(color, srcnn.scale)
        #bad_color = shave(bad_color, padding)

        img_y = image[:, :, 0]
        label, bicubic = preprocess(img_y, srcnn.scale)

        bicubic = np.reshape(bicubic, (1, bicubic.shape[0], bicubic.shape[1], 1))
        out = (srcnn.model(bicubic)).numpy()
        out = out[0, :, :, 0]
        out = out[:, :]*255
        out[out[:]>255] = 255
        out[out[:]<0] = 0
        out = out.astype(np.uint8)
        out = shave(out, srcnn.scale)
        out = np.reshape(out, (out.shape[0], out.shape[1], 1))

        label = label[:, :]*255
        label[label[:, :]>255] = 255
        label[label[:, :]<0] = 0
        label = label.astype(np.uint8)
        label = shave(label, srcnn.scale)
        label = np.reshape(label, (label.shape[0], label.shape[1], 1))
        
        bicubic = np.reshape(bicubic, (bicubic.shape[1], bicubic.shape[2]))
        bicubic = bicubic[:, :]*255
        bicubic[bicubic[:, :]>255] = 255
        bicubic[bicubic[:, :]<0] = 0
        bicubic = bicubic.astype(np.uint8)
        bicubic = shave(bicubic, srcnn.scale)
        bicubic = np.reshape(bicubic, (bicubic.shape[0], bicubic.shape[1], 1))

        origin = np.concatenate((label, color), axis=2)
        origin = cv2.cvtColor(origin, cv2.COLOR_YCrCb2RGB)
        sr_img = np.concatenate((out, color), axis=2)
        sr_img = cv2.cvtColor(sr_img, cv2.COLOR_YCrCb2RGB)
        lr_img = np.concatenate((bicubic, color), axis=2)
        lr_img = cv2.cvtColor(lr_img, cv2.COLOR_YCrCb2RGB)


        psnr_srcnn += psnr_2(origin, sr_img)
        psnr_bicubic += psnr_2(origin, lr_img)

        list1.append(psnr_2(origin, sr_img))
        list2.append(psnr_2(origin, lr_img))
        srcnn_ssim.append(ssim(origin, sr_img, multichannel=True))
        bicubic_ssim.append(ssim(origin, lr_img, multichannel=True))


    print("Mean srcnn psnr " + str(psnr_srcnn / len(os.listdir(path))))
    print("Mean bicubic psnr " + str(psnr_bicubic / len(os.listdir(path))))

    return list1, list2, srcnn_ssim, bicubic_ssim
```

```python
class PSNR_callback(Callback):
    def __init__(self, max_psnr, scale, name_model, psnr_list):
        super(PSNR_callback, self).__init__()
        self.max_psnr = max_psnr
        self.scale = scale
        self.name_model = name_model
        self.psnr = psnr_list[:]

    def on_epoch_end(self, epoch, logs=None):
        psnr_srcnn = 0
        psnr_bicubic = 0
        for file in os.listdir(set5_path):
             image = cv2.imread(os.path.join(set5_path, file))
             image = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
             label, bicubic = preprocess(image[:, :, 0], self.scale)
             bicubic = np.reshape(bicubic, (1, bicubic.shape[0], bicubic.shape[1], 1))
             out = (self.model(bicubic)).numpy()
             out = out[0, :, :, 0]
             out = out[:, :] * 255
             out[out[:, :]>255] = 255
             out[out[:, :] < 0] = 0
             out = out.astype(np.uint8)
             out = out[self.scale:-self.scale, self.scale:-self.scale]
             label = label[:, :]*255
             label[label[:, :]>255] = 255
             label[label[:, :]<0] = 0
             label = label.astype(np.uint8)
             label = label[self.scale:-self.scale, self.scale:-self.scale]
             bicubic = np.reshape(bicubic, (bicubic.shape[1], bicubic.shape[2]))
             bicubic = bicubic[:, :] * 255
             bicubic[bicubic[:, :]>255] = 255
             bicubic[bicubic[:, :]<0] = 0
             bicubic = bicubic.astype(np.uint8)
             bicubic = bicubic[self.scale:-self.scale, self.scale:-self.scale]
             psnr_srcnn += psnr_2(label, out)
             psnr_bicubic += psnr_2(label, bicubic)
        self.psnr.append(psnr_srcnn/len(os.listdir(set5_path)))
        print("Mean PSNR for epoch: %.3f" % (psnr_srcnn/len(os.listdir(set5_path))))
        if psnr_srcnn/len(os.listdir(set5_path)) > self.max_psnr:
            self.max_psnr = psnr_srcnn/len(os.listdir(set5_path))
            if self.name_model == "915":
               if self.scale == 3:
                  self.model.save("best_model_scale3_915.h5")
               elif self.scale == 2:
                  self.model.save("best_model_scale2_915.h5")
               else:
                  self.model.save("best_model_scale4_915.h5")
            elif self.name_model == "935":
                 if self.scale == 3:
                   self.model.save("best_model_scale3_935.h5")
                 elif self.scale == 2:
                   self.model.save("best_model_scale2_935.h5")
                 else:
                   self.model.save("best_model_scale4_935.h5")
            elif self.name_model == "9115":
                 if self.scale == 3:
                    self.model.save("best_model_scale3_9115.h5")
                 elif self.scale == 2:
                    self.model.save("best_model_scale2_9115.h5")
                 else:
                    self.model.save("best_model_scale4_9115.h5")
            elif self.name_model == "9315":
                 if self.scale == 3:
                    self.model.save("best_model_scale3_9315.h5")
                 elif self.scale == 2:
                    self.model.save("best_model_scale2_9315.h5")
                 else:
                    self.model.save("best_model_scale4_9315.h5")
            else:
                 if self.scale == 3:
                   self.model.save("best_model_scale3_955.h5")
                 elif self.scale == 2:
                   self.model.save("best_model_scale2_955.h5")
                 else:
                   self.model.save("best_model_scale4_955.h5")
    
    def on_train_end(self, logs=None):
        plt.plot(self.psnr, "b")
        plt.xlabel("Epochs")
        plt.ylabel("PSNR")
        plt.title("Mean PSNR")
        plt.show()             
```
